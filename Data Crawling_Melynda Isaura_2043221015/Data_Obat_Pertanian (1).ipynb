# MULTI-PAGE SCRAPING

```python
from bs4 import BeautifulSoup
import requests


Products = []
Prices = []

for index in range(1, 21):
    page = requests.get(f'https://bibitonline.com/kategori/perlengkapan/obat-pertanian/page/{index}/')
    soup = BeautifulSoup(page.content, 'html.parser')
    html_Products = soup.find_all('p', class_='name product-title')
    html_Prices = soup.find_all('span', class_='price')

    for each_html_Product in html_Products:
        Products.append(each_html_Product.text)

    for each_html_Price in html_Prices:
        Prices.append(each_html_Price.text)
```

```python
data = {
    'Nama_Produk' : Products,
    'Harga_Produk' : Prices
}
```

```python
pd.DataFrame(data).to_csv('kutip_2.csv')
```
