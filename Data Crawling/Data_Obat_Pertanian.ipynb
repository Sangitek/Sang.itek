```python
html_Nama_Produk = soup.find_all('p', class_='name product-title')
```

```python
Nama_produk = []
for each_html_Nama_Produk in html_Nama_Produk:
    print(each_html_Nama_Produk.text)
    Nama_produk.append(each_html_Nama_Produk.text)
```

```python
html_Harga_Produk = soup.find_all('span', class_='price')
```

```python
Harga_produk = []
for each_html_Harga_Produk in html_Harga_Produk:
    print(each_html_Harga_Produk.text)
    Harga_produk.append(each_html_Harga_Produk.text)
```

```python
data = {
    'Nama_Produk' : Nama_produk,
    'Harga_Produk' : Harga_produk,
}
```

```python
pd.DataFrame(data).to_csv('kutip.csv')
```
